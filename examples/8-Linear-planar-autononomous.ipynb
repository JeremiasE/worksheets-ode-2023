```python
import IPython.display as disp
import sympy
import vector_fields
from sympy import init_printing
init_printing() 
```

```python
(t,x,y)=sympy.var("t,x,y")
```

```python
vector = sympy.Matrix
matrix = sympy.Matrix
```

```python
def normalize(f):
    norm = sympy.sqrt(f[0]**2+f[1]**2)
    return f/norm
```

```python
def plot_linear_system(A,color="#267ab2",show_vars=True):
    f = A*vector([x,y])
    ax = vector_fields.plot_vector_field(normalize(f),(x,-8,8),(y,-8,8),color="lightgrey")
    vector_fields.plot_streamlines(f, (x,-8,8), (y,-8,8),color=color,ax = ax)
    ax.set_aspect('equal')
    if show_vars:
        disp.display(disp.Math(f'$$A= {sympy.latex(A)}$$'))
        disp.display(disp.Math(f'$$e^{{tA}}= {sympy.latex(sympy.simplify(sympy.exp(t*A)))}$$'))
        disp.display(disp.Math(f'$$\sigma(A)= \\{{{sympy.latex(str(A.eigenvals(simplify=True, multiple=True))[1:-1])}\\}}$$'))
    ax.set_xlabel('$x$')
    return ax

def visualize_linear_system(A):
    f = A*vector([x,y])
    return vector_fields.field_player(f)
```

# Phasenportäts linearer autonomer planarer Differentialgleichungen

Im folgenden stellen wir stets Phasenkurven bzw. Trajektorien (also Teilmengen der Ebene, die durch Lösungen der DGL parametrisiert sind) der DGL 
$$\begin{pmatrix} \dot x(t) \\ \dot y(t) \end{pmatrix} = A \begin{pmatrix} x(t) \\ y(t) \end{pmatrix}$$ dar. 

## Instabiler Knoten

```python
A = matrix([[1,0],[0,1]])
plot_linear_system(A);
```

```python
A = matrix([[2,0],[0,1]])
plot_linear_system(A);
```

```python
A = matrix([[1,1],[0,1]])
plot_linear_system(A);
```

## Stabiler Knoten

```python
A = matrix([[-2,0],[0,-1]])
plot_linear_system(A);
```

```python
A = matrix([[-1,1],[0,-1]])
plot_linear_system(A);
```

## Sattel

```python
A = matrix([[1,0],[0,-1]])
plot_linear_system(A);
```

```python
C = matrix([[1,0],[0,-1]])
P = matrix([[1,0],[1,1]])
plot_linear_system(P*C*P**(-1));
```
