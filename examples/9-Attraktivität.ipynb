```python
import IPython.display as disp
import sympy
import vector_fields
from sympy import init_printing
init_printing() 
```

```python
(t,x,y)=sympy.var("t,x,y")
r=sympy.sqrt(x**2+y**2)
```

# Eine Differentialgleichung mit instabiler aber attraktiver Ruhelage

```python
vector_fields.plot_streamlines((-y/2*(1-x/r)+x*(1-r),(x/2*(1-x/r)+y*(1-r))),(x,-6,6),(y,-4,4)); 
```

# Eine Differentialgleichung mit instabiler und global attraktiver Ruhelage

```python
vector_fields.plot_streamlines(((x**2*(y-x)+y**5)/(r**2*(1+r**4)),(y**2*(y-2*x))/(r**2*(1+r**4))),(x,-6,6),(y,-4,4));
```
